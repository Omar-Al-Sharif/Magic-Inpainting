```python
from tkinter import *
from tkinter import filedialog
from PIL import Image, ImageTk, ImageDraw
import os

line_start = None
lines = []
imageWidth=None
imageLength=None

def save_image():
    global image, original_width, original_height

    draw = ImageDraw.Draw(image)

    for i in range(0, len(lines), 4):
        x0, y0, x1, y1 = lines[i]
        draw.rectangle((x0, y0, x1, y1), outline="black", width=12)

    desktop_path = os.path.join(os.path.expanduser("~"), "Desktop")

    file_path = filedialog.asksaveasfilename(defaultextension=".png", filetypes=[("PNG files", "*.png"), ("All files", "*.*")], initialdir=desktop_path)

    if file_path:
        # Rescale the image back to its original size before saving
        resized_image = image.resize((imageWidth, imageLength), Image.LANCZOS)
        resized_image.save(file_path)
        # # Keep a reference to the PhotoImage object
        # app.image = ImageTk.PhotoImage(resized_image)
        # canvas.create_image(0, 0, image=app.image, anchor='nw')

def get_x_and_y(event):
    global lasx, lasy
    lasx, lasy = event.x, event.y
    
def draw_line(event):
    global line_start

    if line_start is None:
        line_start = (event.x, event.y)
    else:
        line_end = (event.x, event.y)

        x0, y0 = canvas.canvasx(line_start[0]), canvas.canvasy(line_start[1])
        x1, y1 = canvas.canvasx(line_end[0]), canvas.canvasy(line_end[1])

        # Draw a rectangle
        canvas.create_rectangle(x0, y0, x1, y1, outline="black", width=12)

        lines.append((x0, y0, x1, y1))
        line_start = None

def draw_smth(event):
    global lasx, lasy
    canvas.create_line((lasx, lasy, event.x, event.y),
                       fill='red',
                       width=2)
    lasx, lasy = event.x, event.y

app = Tk()
app.geometry("800x800")
canvas = Canvas(app, bg='black')
canvas.pack(anchor='nw', fill='both', expand=1)
full_path = filedialog.askopenfilename(parent=app, initialdir=".", title='Choose an image.')
image = Image.open(full_path)
imageWidth, imageLength = image.size 
image = image.resize((800, 800), Image.LANCZOS)
app.image = ImageTk.PhotoImage(image)
canvas.create_image(0, 0, image=app.image, anchor='nw')

canvas.bind("<Button-1>", get_x_and_y)
canvas.bind("<B1-Motion>", draw_smth)
# canvas.bind("<Button 1>", draw_line)

save_button = Button(app, text="Save Image", command=save_image)
save_button.pack(side=BOTTOM)

app.mainloop()
```
